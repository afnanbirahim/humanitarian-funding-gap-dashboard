```python
# ==========================================
# MINIMAL CLEAN OECD DAC2A CSV FOR POWER BI
# Best for dashboarding
# ==========================================

import pandas as pd
from google.colab import files

print("Upload your OECD DAC2A CSV file now...")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

clean_df = df[[
    'Donor',
    'Recipient',
    'TIME_PERIOD',
    'OBS_VALUE',
    'Measure',
    'Price base'
]].copy()

clean_df.rename(columns={
    'TIME_PERIOD': 'Year',
    'OBS_VALUE': 'Aid_USD_Millions',
    'Price base': 'Price_Base'
}, inplace=True)

# Clean text
for col in ['Donor', 'Recipient', 'Measure', 'Price_Base']:
    clean_df[col] = clean_df[col].astype(str).str.strip()

# Convert numeric fields
clean_df['Year'] = pd.to_numeric(clean_df['Year'], errors='coerce')
clean_df['Aid_USD_Millions'] = pd.to_numeric(clean_df['Aid_USD_Millions'], errors='coerce')

# Remove bad rows
clean_df = clean_df.dropna(subset=['Donor', 'Year', 'Aid_USD_Millions'])

# Convert year to integer
clean_df['Year'] = clean_df['Year'].astype(int)

# Remove duplicates
clean_df = clean_df.drop_duplicates()

# Sort
clean_df = clean_df.sort_values(['Donor', 'Year']).reset_index(drop=True)

print("Final shape:", clean_df.shape)
print(clean_df.head(10))

output_file = "dac_donor_oda_clean_minimal.csv"
clean_df.to_csv(output_file, index=False)

files.download(output_file)
```
